# Popularity Based Recommendation System
formula to select top 50 books which books have vote more than 250

```python
ratings_withs_name = ratings.merge(books, on='ISBN')
```

```python
ratings_withs_name
```

```python
num_rating_df = ratings_withs_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'}, inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_withs_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'}, inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on='Book-Title')
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings', ascending=False).head(50)
```

```python
popular_df.shape
```

```python
books.head()
```

```python
popular_df.shape
```

```python
popular_df = popular_df.merge(books, on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Year-Of-Publication','Image-URL-M','num_ratings','avg_ratings']]
popular_df.shape
```

# Collaborative Filtering Based Recommendation System
user has given min 200 ratings vote

```python
# user given greater than 200 rating 
x = ratings_withs_name.groupby(by='User-ID').count()['Book-Rating'] > 200
user_has_given_ratings = x[x].index
```

```python
user_has_given_ratings
```

```python
filtered_ratings = ratings_withs_name[ratings_withs_name['User-ID'].isin(user_has_given_ratings)]
```

```python
y = filtered_ratings.groupby(by='Book-Title').count()['Book-Rating']>=50
famous_books =  y[y].index
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
final_ratings.duplicated().sum()
```

```python
pt = final_ratings.pivot_table(index='Book-Title', columns='User-ID', values='Book-Rating')
```

```python
pt.fillna(0, inplace=True)
```

```python
pt = pt.astype('int32')
```

```python
pt.shape
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pt)
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])), key=lambda x:x[-1], reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    return data
```
